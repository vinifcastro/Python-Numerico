# Métodos Iterativos para Cálculo de Raízes

## Raiz Quadrada

O método iterativo clássico para calcular a raiz quadrada de um número real positivo \(a\) é dado por:

$$
x_{1} = c, \quad 
x_{n+1} = \frac{1}{2}\left(x_{n} + \frac{a}{x_{n}}\right), \quad n \geq 1.
$$

---

## Raiz \(k\)-ésima (Método de Newton-Raphson)

Para calcular a raiz \(k\)-ésima de um número real positivo \(a\), aplicamos o método de Newton-Raphson à função \(f(x) = x^k - a\), obtendo a iteração:

$$
x_{1} = c, \quad 
x_{n+1} = \frac{(k-1)x_{n} + \tfrac{a}{x_{n}^{k-1}}}{k}, \quad n \geq 1.
$$


```python
import math
import matplotlib.pyplot as plt
```

```python
def raiz_quadrada(a, c, max_iter=20, tol=1e-10):
    x = c
    seq = [x]
    for _ in range(max_iter):
        x_next = 0.5 * (x + a / x)
        seq.append(x_next)
        if abs(x_next - x) < tol:
            break
        x = x_next
    return seq
```

```python
# Exemplo: sqrt(10), chute inicial c=1
a = 10
seq_sqrt = raiz_quadrada(a, c=1)

print("Aproximações para sqrt(10):")
for i, val in enumerate(seq_sqrt):
    print(f"Iteração {i}: {val}")

print(f"\nValor real com math.sqrt: {math.sqrt(a)}")
```

```python
plt.plot(seq_sqrt, marker="o")
plt.axhline(math.sqrt(a), color="r", linestyle="--", label="Valor real")
plt.title("Convergência para sqrt(10)")
plt.xlabel("Iteração")
plt.ylabel("Aproximação")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def raiz_kesima(a, k, c, max_iter=20, tol=1e-10):
    x = c
    seq = [x]
    for _ in range(max_iter):
        x_next = ((k - 1) * x + a / (x**(k-1))) / k
        seq.append(x_next)
        if abs(x_next - x) < tol:
            break
        x = x_next
    return seq
```

```python
# Exemplo: raiz cúbica de 10 (k=3), chute inicial c=1
a = 10
k = 3
seq_cubic = raiz_kesima(a, k, c=1)

print(f"Aproximações para raiz {k}-ésima de {a}:")
for i, val in enumerate(seq_cubic):
    print(f"Iteração {i}: {val}")

print(f"\nValor real com math.pow: {a**(1/k)}")
```

```python
plt.plot(seq_cubic, marker="o", color="g")
plt.axhline(a**(1/k), color="r", linestyle="--", label="Valor real")
plt.title(f"Convergência para raiz {k}-ésima de {a}")
plt.xlabel("Iteração")
plt.ylabel("Aproximação")
plt.legend()
plt.grid(True)
plt.show()
```
